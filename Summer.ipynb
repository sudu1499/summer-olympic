# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
data=pd.read_csv("summer.csv")
```

### 1. In how many cities Summer Olympics is held so far?

```python
len(data.groupby('City'))
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
k=[]
j=data[data['Medal']=='Gold']
for i in data['Sport'].unique():
    k.append([i,len(j[j['Sport']==i])])
k=pd.DataFrame(k,columns=['sport','number of gold']).sort_values(by='number of gold',ascending=False).head()
print(k.iloc[:1,:].values[0][0])
k.plot(x='sport',y='number of gold',kind='bar')
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
k=[]
for i in data['Sport'].unique():
    k.append([i,len(data[data['Sport']==i])])
k=pd.DataFrame(k,columns=['sport','medal count']).sort_values(by='medal count',ascending=False).head()
print(k.iloc[:1,:].values[0][0])
k.plot(x='sport',y='medal count',kind='bar')
```

### 4. Which player has won most number of medals? (Top 5)

```python
k=[]
for i,j in data.groupby('Athlete'):
    k.append([i,len(j)])
k=pd.DataFrame(k,columns=['Athlete','medal count']).sort_values(by='medal count',ascending=False).head()
print(k.iloc[:1,:].values[0][0])
k.plot(x='Athlete',y='medal count',kind='bar')
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
g=data[data['Medal']=='Gold']
k=[]
for i,j in g.groupby('Athlete'):
    k.append([i,len(j)])
k=pd.DataFrame(k,columns=['athlete','golds']).sort_values(by='golds',ascending=False).head()
print(k.iloc[:1,:].values[0][0])
k.plot(x='athlete',y='golds',kind='bar')
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
ind=data[data['Country']=='IND']
ind[ind['Medal']=='Gold']['Year'].values[0]
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
k=[]
for i,j in data.groupby('Event'):
    k.append([i,len(j)])
k=pd.DataFrame(k,columns=['event','players count']).sort_values(by='players count',ascending=False).head()
print(k.iloc[:1,:].values[0][0])
k.plot(x='event',y='players count',kind='bar')
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
k=[]
for i in data['Sport'].unique():
    t=data[data['Sport']==i]
    tt=t[t['Gender']=='Women']
    k.append([i,len(tt[tt['Medal']=='Gold'])])
k=pd.DataFrame(k,columns=['Sport','female gold medalist']).sort_values(by='female gold medalist',ascending=False).head()
print(k.iloc[:1,:].values[0][0])
k.plot(x='Sport',y='female gold medalist',kind='bar')
```
